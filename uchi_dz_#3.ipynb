```python
import pandas as pd
from matplotlib import pyplot as plt


%matplotlib inline
```

```python
class BaseGraph:
    def __init__(self, data,name,ax,ay,delimiter,to_float=True):
        """
        data: dataframe or file name  
        name: title
        ax: axis name
        ay: axis name
        delimiter: 
        to_float: convert to float or not
        """
        if type(data) == str:
            self.data = pd.read_csv(data,delimiter=delimiter)
        else:
            self.data =data
        
        if to_float:
            self.data = self.data.astype(float)
        self.name = name
        self.axes = [ax,ay]
        self.show()
        
    def plot(self,x=None,y=None,kind="line"):
        """
        x: dataframe x column name
        y: dataframe x column name
        kind:
            ‘line’ : line plot (default)
            ‘bar’ : vertical bar plot
            ‘barh’ : horizontal bar plot
            ‘hist’ : histogram
            ‘box’ : boxplot
            ‘kde’ : Kernel Density Estimation plot
            ‘density’ : same as ‘kde’
            ‘area’ : area plot
            ‘pie’ : pie plot
            ‘scatter’ : scatter plot
            ‘hexbin’ : hexbin plot
        """
        
        if x!=None and y!=None:
            ax = self.data.plot(title=self.name,kind = kind,x=x,y=y)
        elif y==None:
            ax = self.data.plot(title=self.name,kind = kind,x=x)
        else:
            ax = self.data.plot(title=self.name,kind = kind)
        
        ax.set_xlabel(self.axes[0])
        ax.set_ylabel(self.axes[1])
    
    def show(self):
        print(self.data)
        print("name = " + self.name,"\naxes_names = " + str(self.axes))
        print("columns = "+str(self.data.columns))
        
class BarGraph(BaseGraph):
    def plot(self,x=None,y=None,kind="bar"):
        """
        x: dataframe x column name
        y: dataframe x column name
        kind:
            ‘line’ : line plot (default)
            ‘bar’ : vertical bar plot
            ‘barh’ : horizontal bar plot
            ‘hist’ : histogram
            ‘box’ : boxplot
            ‘kde’ : Kernel Density Estimation plot
            ‘density’ : same as ‘kde’
            ‘area’ : area plot
            ‘pie’ : pie plot
            ‘scatter’ : scatter plot
            ‘hexbin’ : hexbin plot
        """
        
        if x!=None and y!=None:
            ax = self.data.plot(title=self.name,kind = kind,x=x,y=y)
        elif y==None:
            ax = self.data.plot(title=self.name,kind = kind,x=x)
        else:
            ax = self.data.plot(title=self.name,kind = kind)
        
        ax.set_xlabel(self.axes[0])
        ax.set_ylabel(self.axes[1])
```
